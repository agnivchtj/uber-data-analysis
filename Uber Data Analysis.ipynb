Uber is a popular American tech company known for ride-hailing, food delivery, package delivery, couriers, freight transportation, electric bicycle and motorized scooter rental via partnership with Lime. Uber does not own any vehicles but takes a commission from every booking on its platform. Fares, which are priced using a dynamic model based on local supply and demand at the time of booking, are quoted to the customer in advance.

As of the fourth quarter of 2021, Uber had 118 million monthly active users worldwide and generated an average of 19 million trips per day. We will be analyzing data pertaining to Uber trips in order to produce key insights and/or statistics (using Python) that can be useful for analysts.

## Import libraries and data

To start off, we will import the required Python libraries for data analysis and also read the trip data into a dataframe which we can work with.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import calendar

data = pd.read_csv('uber_drives.csv')
print(data)
```

## Characteristics of the dataset

First, we can look to analyze the basic features of our dataset, such as the number of rows, columns and column names as well as missing values in the dataset that we have to consider later on. We can also look at the category names, purposes, start points and destinations for Uber trips.

```python
no_of_categories = len(data.columns)
print('Number of columns in this dataset: ', no_of_categories)
print('The columns are following: ', list(data.columns))

data_length = len(data)
print('Number of rows in this dataset: ', data_length)

no_of_missing_values = data.isnull().any(axis = 1).sum()
print('Number of rows w. at least one missing value: ', no_of_missing_values)
print('\n')

## We can rename 'START*' and 'STOP*' columns
data.rename(columns = {'START*': 'SOURCE', 'STOP*': 'DESTINATION'}, inplace=True)

## CATEGORIES
categories = data['CATEGORY*'].unique()
print('The categories for Uber trips are: ', categories.tolist())

## PURPOSES
purposes = data['PURPOSE*'].unique()
print('The purposes for Uber trips are: ', purposes.tolist())

## LOCATIONS
no_of_start_loc = len(data['SOURCE'].unique())
print('The number of start locations for Uber trips are: ', no_of_start_loc)

no_of_stop_loc = len(data['DESTINATION'].unique())
print('The number of stop locations for Uber trips are: ', no_of_stop_loc)
```

## Data cleaning and filtering

Since there are missing figures from the dataset which can affect our analysis, we need to filter some rows from our dataset. The only column which should have any NaNs should be the 'Purpose of trip' column (which isn't always clear) and all other columns should contain data which we can make use of.

```python
data.dropna(subset=['START_DATE*', 'END_DATE*', 'CATEGORY*', 'SOURCE', 'DESTINATION', 'MILES*'], inplace=True)
print(data)
```

## Data processing

Now that we have filtered missing values, we can use the data in our dataset to find some insights on the following:
- How long are Uber trips?
- How far do people travel with Uber?
- What is the most visited destination on these trips?
- What is the most popular reason, or purpose, for taking an Uber?
- Which days in the year have had the most & least number of Uber trips?

```python
## Convert dates to timestamps
data['START_DATE*'] = pd.to_datetime(data['START_DATE*'])
data['END_DATE*'] = pd.to_datetime(data['END_DATE*'])
print(data)
print('\n')

## How long do people travel with Uber?
trip_lengths = data['END_DATE*'] - data['START_DATE*']
data['TIME_DURATION'] = trip_lengths

avg_trip_length = data['TIME_DURATION'].mean()
print('The average duration of an Uber ride: ', avg_trip_length)

## How far do people travel with Uber?
avg_trip_distance = data['MILES*'].mean()
print('The average distance travelled in an Uber ride: %d miles' % (avg_trip_distance))

## Most Frequent Destination
most_frequent_destination = data['DESTINATION'].value_counts().index[0]
no_of_trips = data['DESTINATION'].value_counts()[0]
print('The most visited destination is %s with %d trips' % (most_frequent_destination, no_of_trips))
print('\n')

## Most Popular Reasons / Purposes
most_frequent_purpose = data['PURPOSE*'].value_counts().index[0]
no_of_trips = data['PURPOSE*'].value_counts()[0]
print('The most popular reason for taking Uber is %s with %d trips' % (most_frequent_purpose, no_of_trips))

second_frequent_purpose = data['PURPOSE*'].value_counts().index[1]
no_of_trips = data['PURPOSE*'].value_counts()[1]
print('The second-most popular reason is %s with %d trips' % (second_frequent_purpose, no_of_trips))

third_frequent_purpose = data['PURPOSE*'].value_counts().index[2]
no_of_trips = data['PURPOSE*'].value_counts()[2]
print('The third-most popular reason is %s with %d trips' % (third_frequent_purpose, no_of_trips))
print('\n')

## Days with Most/Least trips
date_with_most_trips = data['START_DATE*'].value_counts().index[0]
print('The date with most number of Uber trips: ', date_with_most_trips)

date_with_least_trips = data['START_DATE*'].value_counts().index[1153]
print('The date with least number of Uber trips: ', date_with_least_trips)
```

## Data visualization

We can also look to plot & visualize data on the number of trips based on purpose. This would allow us to see which reasons are more popular for taking an Uber and why, e.g. it could be that Uber might not be a popular option for commuting regularly (as people may live close to their workplace or prefer to walk, bike or use public transport) but may be more popular when stepping out for a meal or outing. 

A similar analysis can be conducted for number of trips based on destination, enabling us to see which cities (and countries by extension) may use Uber more. We can keep this to a top 15 in order to keep the visualization concise and meaningful as well.

```python
# Purpose Frequencies
purposes = list(data['PURPOSE*'].value_counts().index)
frequencies = list(data['PURPOSE*'].value_counts())
mf_df = pd.DataFrame(list(zip(purposes, frequencies)), columns = ['Purpose', 'Number of trips'])
mf_df.plot(x = 'Purpose', y = 'Number of trips', title = 'Most Popular Purposes for taking an Uber', kind = 'bar', figsize=(20,5))
plt.show()
```

```python
## Destination Frequencies
dest_overview = pd.DataFrame(data['DESTINATION'].value_counts())
dest_overview.drop('Unknown Location', axis=0, inplace=True)

destinations = dest_overview.head(15).index.values
frequencies = list(dest_overview.head(15)['DESTINATION'])
df_df = pd.DataFrame(list(zip(destinations, frequencies)), columns = ['Destination', 'Number of trips'])
df_df.plot(x = 'Destination', y = 'Number of trips', title = 'Most Popular Destinations w/ Uber', kind = 'bar', figsize=(20,5))
plt.show()
```

## Trip frequency

We also would like to know some things like the number of trips per day and per month in order to have timeframed graphs. This serves to visualize trip frequency which would provide us with an accurate overview of demand for Uber rides and prove useful for identifying times of the year, months of the year etc. when there can be peak traffic on the app. This would affect how fares are priced for customers, how compensation is calculated for drivers etc.

- What are the number of trips per day? Days of the year 2016
- What are the number of trips per month? Months range from Jan, Feb, ..., Dec

```python
data['START_TIME'] = data['START_DATE*'].dt.time
data['END_TIME'] = data['END_DATE*'].dt.time
data['MONTH'] = data['START_DATE*'].dt.month
data['START_DATE*'] = data['START_DATE*'].dt.date
data['END_DATE*'] = data['END_DATE*'].dt.date
print(data)
```

```python
daily_data = data['START_DATE*'].value_counts().sort_index()

days = [str(x) for x in list(daily_data.index)]
no_of_trips = list(daily_data)
date_df = pd.DataFrame(list(zip(days, no_of_trips)), columns = ['Date', 'Number of trips'])
date_df.plot(x = 'Date', y = 'Number of trips', title = 'Daily Trip Data', kind = 'line', figsize=(20,5))
plt.show()
```

```python
monthly_data = data['MONTH'].value_counts().sort_index()

months = [calendar.month_abbr[x] for x in list(monthly_data.index)]
no_of_trips = list(monthly_data)
month_df = pd.DataFrame(list(zip(months, no_of_trips)), columns = ['Month', 'Number of trips'])
month_df.plot(x = 'Month', y = 'Number of trips', title = 'Monthly Trip Data', kind = 'line', figsize=(20,5))
plt.show()
```
